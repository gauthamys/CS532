```python
import pandas as pd
```

```python
df = pd.read_csv('simulated_qpu_data_enhanced.csv')
```

```python
cost_atom = 0.01
cost_photon = 0.05
cost_spin = 0.20

execution_costs = []
```

```python
for i in range(len(df)):
    if i == 0:
        # Day 1: only new workloads
        new_total = df.loc[i, 'new_blocks_total']
        new_atom = df.loc[i, 'new_blocks_atom']
        new_photon = df.loc[i, 'new_blocks_photon']
        new_spin = df.loc[i, 'new_blocks_spin']
        w_new = df.loc[i, 'workloads_new_blocks']

        cost = (
            (new_atom / new_total) * w_new * cost_atom +
            (new_photon / new_total) * w_new * cost_photon +
            (new_spin / new_total) * w_new * cost_spin
        )
        print(f"Day {i + 1}: Cost = {cost:.4f}")  # Print the cost for day 1 for clarity
        execution_costs.append(cost)
    else:
        # --- New block workload cost ---
        new_total = df.loc[i, 'new_blocks_total']
        new_atom = df.loc[i, 'new_blocks_atom']
        new_photon = df.loc[i, 'new_blocks_photon']
        new_spin = df.loc[i, 'new_blocks_spin']
        w_new = df.loc[i, 'workloads_new_blocks']

        cost_new = (
            (new_atom / new_total) * w_new * cost_atom +
            (new_photon / new_total) * w_new * cost_photon +
            (new_spin / new_total) * w_new * cost_spin
        )

        # --- Older block workload cost ---
        w_old = df.loc[i, 'workloads_older_blocks']
        cumulative_blocks_prev = df.loc[i - 1, 'cumulative_blocks']
        
        # Sum of blocks from previous days
        old_atom = df.loc[:i - 1, 'new_blocks_atom'].sum()
        old_photon = df.loc[:i - 1, 'new_blocks_photon'].sum()
        old_spin = df.loc[:i - 1, 'new_blocks_spin'].sum()

        cost_old = (
            (old_atom / cumulative_blocks_prev) * w_old * cost_atom +
            (old_photon / cumulative_blocks_prev) * w_old * cost_photon +
            (old_spin / cumulative_blocks_prev) * w_old * cost_spin
        )

        total_cost = cost_new + cost_old
        print(f"Day {i + 1}: Cost = {total_cost:.4f}")
        execution_costs.append(total_cost)
```

```python
df['workload_execution_cost'] = execution_costs
```

```python
# Save the DataFrame to a CSV file
df.to_csv('simulation_with_costs.csv', index=False)
```
